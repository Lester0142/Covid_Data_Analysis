```python
import pandas as pd
from fbprophet import Prophet
import warnings
import logging
import plotly
logging.getLogger('fbprophet').setLevel(logging.WARNING)
import plotly.graph_objs as go
import plotly.offline as py
from plotly.offline import init_notebook_mode
```

```python
Weekly_df = pd.read_csv("Dataset\\Week_Death_Vac_2Y.csv")
```

```python
#https://dev.to/yash_makan/how-to-predict-future-using-python--3i39
#https://facebook.github.io/prophet/docs/quick_start.html
```

## Weekly Deaths

```python
m = Prophet()
```

```python
Weekly_df['ds'] = Weekly_df['Date']
Weekly_df['y'] = Weekly_df['Weekly Deaths']
```

```python
m = Prophet()
model = m.fit(Weekly_df)
```

```python
future = m.make_future_dataframe(periods=180)
forecast = m.predict(future)
```

```python
df_forecast = forecast.copy()
df0 = Weekly_df.copy()

trace1 = go.Scatter(
    name = 'trend',
    mode = 'lines',
    x = list(df_forecast['ds']),
    y = list(df_forecast['yhat']),
    marker=dict(
        color='red',
        line=dict(width=3)
    )
)

tracex = go.Scatter(
    name = 'Actual Deaths',
   mode = 'markers',
   x = list(Weekly_df['Date']),
   y = list(Weekly_df['Weekly Deaths']),
   marker=dict(
      color='black',
      line=dict(width=2)
   )
)
data = [trace1,tracex]

layout = dict(title=f'Weekly Deaths Prediction',
             xaxis=dict(title = 'Dates', ticklen=2, zeroline=True))

figure=dict(data=data,layout=layout)
py.offline.iplot(figure)
```
